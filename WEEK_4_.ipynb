L4.5 The Obvious Sort

```python
l=[12,10,7,18,6,43,8,35]
l.sort()
print(l)
```

```python
l=[12,10,7,18,6,43,8,35]
x=[]

while(len(l)>0):
  min=l[0]
  for i in range(len(l)):
     if(l[i]<min):
        min=l[i]
  x.append(min)
  l.remove(min)

print(l)
print(x)
```

L4.7 Matrix Addition

```python
#Matrix by frist principles

dim=3

r1=[1,2,3]
r2=[4,5,6]
r3=[7,8,9]

s1=[1,2,1]
s2=[6,2,3]
s3=[4,2,1]

A=[]
A.append(r1)
A.append(r2)
A.append(r3)

B=[]
B.append(s1)
B.append(s2)
B.append(s3)

print(A)
print(B)

#i nee to add A and B

C=[[0,0,0],[0,0,0],[0,0,0]]
D=[[0,0,0],[0,0,0],[0,0,0]]

#C=[[],[],[]]
for i in range(dim):
    for j in range(dim):
      C[i][j]=A[i][j]+B[i][j]  #Matrix Addition
      D[i][j]=A[i][j]*B[i][j]   #Matrix Multiplication


print(C)
print(D)
```

L4.8-9| Matrix Multiplication 1 and 2

```python
r1=[1,2,3]
r2=[4,5,6]
r3=[7,8,9]

s1=[1,2,1]
s2=[6,2,3]
s3=[4,2,1]

A=[]
B=[]
A.append(r1)
A.append(r2)
A.append(r3)

B.append(s1)
B.append(s2)
B.append(s3)

C=[[0,0,0],[0,0,0],[0,0,0]]

dim=3

#C C[i][j] is the dot product of the ith row of A and the jth column of B

for i in range(dim):
    for j in range(dim):
            for k in range(dim):
               C[i][j] = C[i][j] + A[i][k]*B[k][j]

print(C)

# C[i][j]=dot product of A[i][....] and B[....][j]

```
