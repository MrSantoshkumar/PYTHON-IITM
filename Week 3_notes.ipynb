#Rounding an integer using Round function.

```python
n = 3.04
print(round(n), type(round(n))) #rounds back to nearest integer. (if no other parameter is mentioned then the class would be <int>)
```

```python
n = 3.04
print(round(n, 0), type(round(n))) #rounds back to nearest integer but the class would be <bool>
```

```python
n = 4748.67489
print(round(n, -1))  #rounds back to nearest ten (type would be <bool>)
```

Similarly can use -2, -3, -4.... for rounding uptill nearest hundred, thousand or ten-thousand. (type would be bool)

Can use 1,2,3.... for rounding uptill first, second or third decimal place. (type would remain bool).

```python
#eg.

n = 4748.67489
print(round(n, 1)) #rounds upto one place after decimal
print(round(n, 2)) #rounds upto two places after decimal
```

```python
#suppose in case of 2.5, 3.5, 4.5, etc. when the number lies exactly in middle of two consecutive integers:
#in such cases (and only in such cases), it gets round off to its nearest EVEN number.

#ie. round(4.5) -> 4, but round(5.5) -> 6

n = 4.5
print(round(n))
```

#Iterable Unpacking

```python
#unpacking
dob =  '29/09/2004'  #dd/mm/yyyy
day, month, year = dob.split('/')  #here we unpacked the list.
print(type(day)) #here they are in <str> data type.
day, month, year = int(day), int(month), int(year)
print(day, month, year)
print(type(day))
```

```python
a, b, c, d = '1234'
print(a, b, c, d)
```

Unpacking can be done with any ITERABLE datatype like strings, lists, tuples, etc.

```python
#extended unpacking

a, b, *c = '12345' #using of '*' before c allows you to store rest of the charecters as seperate elements of a list.
print(a, b, c)
```

```python
first, *middle, last = '1234'
print(first, middle, last) #first and last are mentioned, the remaining ones will be stored in middle as seperate elements of a list.
```

#for loop

```python
#for loop with range:

for i in range(10): #will iterate for 10 times, the value of i = 0, 1, 2,..., 9
  print(i)
```

```python
#parameters fo range(start, end, step)

#eg. print first 5 even numbers

for i in range(2, 11, 2): #i will start from 2, will go till 10(one less than 11) and will take 2 steps(means would skip one number)
  #i = 2, 4, 6, 8, 10
  print(i)
```

```python
#for loop without range: foreach

#suppose you have any iterable object like string or list, but you want to iterate through each and every element of that object.
#then you should use for loop without range.

s = 'testcase'
for char in s:
  print(char) #it will print each charecter of string 's' in a seperate line.
```

#for or while (where to use which loop)

If you're unsure about the number of iterations then you should use while loop.

```python
#eg. accept numbers from user until you encounter 'END' and print sum of all those integers.

#here we're unsure about the number of times we'll have to accept the input, thus we should use while loop here.

sum = 0
n = input()

while n != 'END': #condition with while loop is opposite of our terminating condition. (This means when we'll encounter terminating condition, we'll come out of the loop).
  sum += int(n)
  n = input()

print(sum)
```

When you're sure about the number of iterations then you should use 'for' loop.

```python
#eg. accept a postitive number as input from the user and print the sum of natural numbers less than or equal to that number.

#basically the ques. is asking to print sum of first n natural numbers, thus we're sure that we'd need 'n' iterations.
#hence for loop is an appropriate choice here.

n = int(input())
sum = 0
for i in range(1, n+1): #starting from 1 and going till n (incuding n, hence we're wriing n+1)
  sum += i

print(sum)
```

When you wish to iterate through each and every element of an iterable object then use for loop without range.

```python
#eg. Without using inbuilt 'in' function, check whether a particular letter is present in a string or not.

word = 'dipanshu'

present = False #concept of flag: we're initially assuming that the letter 'a' is not present in the word.
#in case we found 'a' in word, we'll change the value of present variable to False.

for letter in word:
  if letter == 'a': #checks that whether any charecter of word is 'a' or not.
    present = True

print(present)
```
